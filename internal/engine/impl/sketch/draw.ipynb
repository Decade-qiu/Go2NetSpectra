```python
import matplotlib.pyplot as plt

# 文件名
filename = "flow.txt"

x = []
y = []

# 读取文件
with open(filename, "r") as f:
    for line in f:
        parts = line.strip().split()
        if len(parts) < 3:
            continue  # skip empty or malformed lines
        true_val = int(parts[1])
        est_val = int(parts[2])
        x.append(true_val)
        y.append(est_val)

# Plot scatter
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', alpha=0.7)
plt.plot([min(x), max(x)], [min(x), max(x)], 'r--', label='y=x')  # reference line
plt.xlabel("True Value")
plt.ylabel("Estimated Value")
plt.title("True vs Estimated Values Scatter Plot")
plt.yscale('log')
plt.xscale('log')
plt.legend()
plt.grid(True)
plt.show()
```
